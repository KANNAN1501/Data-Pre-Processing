```python
import pandas as pd
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/diabetes.csv"
df = pd.read_csv(url)
df.to_csv("DiabetesData.csv", index=False)
df = pd.DataFrame(df)
df
```

```python
df = df.rename(columns= {"No_Pation" : "Patients"})
df
```

```python
df["Gender"].unique()
```

```python
df["Gender"] = df['Gender'].str.strip().str.upper()
df['Gender']
```

```python
df['CLASS'] = df['CLASS'].str.strip().str.upper()
```

```python
df['CLASS'].unique()
```

```python
df['Gender'].unique()
```

```python
df = df.dropna(subset=['Gender'])
df
```

```python
df['Gender'].unique()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
num = df.select_dtypes(include=["number"]).drop(columns=['ID'], errors = "ignore")
plt.figure(figsize=(12,6))
sns.boxplot(data=num)
plt.title("Box Plot")
plt.xticks(rotation= 45)
plt.grid(True)
plt.show()
```

```python
df.isnull().sum()
```

```python
df.fillna({'AGE' : df['AGE'].median(),
           'Urea': df['Urea'].median(),
           'Cr' : df['Cr'].mean(),
           'HbA1c': df['HbA1c'].mean(),
           'Chol' : df['Chol'].mean(),
           'TG' : df['TG'].median(),
           'HDL' : df['HDL'].mean(),
           'LDL' : df['LDL'].mean(),
           'VLDL' : df['VLDL'].median(),
           'BMI' : df['BMI'].mean()}, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
cr_threshold = df['Cr'].quantile(0.995)
urea_threshold = df['Urea'].quantile(0.999)
df = df[df['Cr'] <= cr_threshold]
df = df[df['Urea'] <= urea_threshold]
cr_threshold
```

```python
import numpy as np 
from scipy import  stats
lipid_Col = ['LDL', 'VLDL', 'HDL', 'TG', 'Chol']
z_score = np.abs(stats.zscore(df[lipid_Col]))
df = df[(z_score < 3).all(axis = 1)]
df
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
lable_encode = LabelEncoder()
df['Gender'] = lable_encode.fit_transform(df['Gender'])
df['Gender'].unique()
```

```python
df['CLASS'].unique()
```

```python
df = pd.get_dummies(df, columns=['CLASS'], prefix='CLASS', drop_first=True)
df.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
normalized = MinMaxScaler()
df = pd.DataFrame(normalized.fit_transform(df),columns=df.columns)
df
```
